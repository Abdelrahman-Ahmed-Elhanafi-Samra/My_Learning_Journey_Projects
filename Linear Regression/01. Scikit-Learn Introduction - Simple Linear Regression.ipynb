## Simple Linear Regression with SK-Learn üß™

````python
# Import linear model
from sklearn import linear_model
````

````python
# Instantiation/create a LinearRegression object
lm_reg = linear_model.LinearRegression()
````

````python
type(lm_reg)
````

````python
X = bost_df['RM']
y = bost_df['Price']
````

````python
# Print the dimensions of X and y before reshaping
print('Dimensions of X before reshaping: ', X.shape)
print('Dimensions of y before reshaping: ', y.shape)
````

````python
# Reshape X and y
X_reshaped = np.array(X).reshape(-1, 1)
y_reshaped = np.array(y).reshape(-1, 1)
````

````python
print('Dimensions of X after reshaping: ', X_reshaped.shape)
print('Dimensions of y after reshaping: ', y_reshaped.shape)
````

## Fitting Linear Regression

````python
# Fit linear model
lm_reg.fit(X_reshaped, y_reshaped)
````

## Check the lm object ‚òëÔ∏è

````python
print([attr for attr in dir(lm_reg) if '_' and '__' not in attr])
````

## Check the Intercept and Coefficients ‚òëÔ∏è

````python
# Intercept
print('The intercept of simple linear regression is ', lm_reg.intercept_)
````

````python
# Coefficients
print('The coefficients of simple linear regression is ', lm_reg.coef_)
````

## Saving the results into a table üíæ

````python
inter_coef_df = pd.DataFrame({'Intercept' : lm_reg.intercept_ ,
                              'Coefficient' : lm_reg.coef_.flatten()})
inter_coef_df
````

## The Goodness-of-Fit ü•á

````python
print("The R^2 of model is: {:0.3f}".format(lm_reg.score(X_reshaped, y_reshaped)))
````

## Prediction üèÜ



 - Syntax: 
```python
   y_pred = lm_reg.predict(new_data)
```

````python
y_pred = lm_reg.predict(X_reshaped) # I using the same data for educational purposes only
````

## Plotting The Results üìäüìù

````python
plt.scatter(X, y, color = 'black', alpha = 0.5)
plt.plot(X_reshaped, y_pred, color = 'gold', linewidth = 3)
plt.show()
````
